```python
# Finding Vanderbilt's school id in schools table.
schools[schools['schoolname']== 'Vanderbilt University']
```

```python
# Subsetting college playing DataFrame to only include players from Vanderbilt.
vanderbilt = college_playing[college_playing['schoolid'] == 'vandy']
```

```python
vanderbilt.head()
```

```python
# Merging with the people DataFrame to keep only players from Vanderbilt.
vanderbilt_players = vanderbilt.merge(people, on = 'playerid')
```

```python
vanderbilt_players.head()
```

```python
vanderbilt_players.info()
```

```python
# Dropping duplicates to get accurate number of players
vanderbilt_players.drop_duplicates(subset = 'playerid', inplace = True)
```

```python
vanderbilt_players.head()
```

```python
# Dropping yearid from college_playing.
vanderbilt_players.drop(['yearid', axis = 1, inplace = True)
```

```python
vanderbilt_players = vanderbilt_players[['playerid', 'namefirst', 'namelast']]
```

```python
vanderbilt_players.head(2)
```

```python
vanderbilt_salaries = vanderbilt_players.merge(salaries, on = 'playerid')
```

```python
vanderbilt_salaries.head()
```

```python
vanderbilt_salaries.info()
```

```python
salary = vanderbilt_salaries.groupby('playerid').sum()
```

```python
salary.head()
```

```python
salary_short = salary['salary']
```

```python
salary_short
```

```python
vandy = vanderbilt_players.merge(salary_short, on = 'playerid')
```

```python
vandy.sort_values('salary', ascending = False)
```

```python
fielding_short = fielding[['playerid', 'pos', 'po']]
```

```python
fielding_short.head()
```

```python
for i, player in fielding_short.iterrows():
    if player['pos'] == 'C':
        player['pos'] = 'Battery'
    elif player['pos'] == 'OF':
        player['pos'] = 'Outfield'
    else:
        player['pos'] = 'Infield'
```
